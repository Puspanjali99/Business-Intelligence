```python
mr_m = pd.read_excel('Exports Merchandise.xlsx',sheet_name='monthly')
mr_y = pd.read_excel('Exports Merchandise.xlsx',sheet_name='annual')
mr_q = pd.read_excel('Exports Merchandise.xlsx',sheet_name='quarterly')
```

```python
mr_y.head()
```

```python
mr_y = mr_y.rename(columns={mr_y.columns[0]: 'Period'}) #name first column as period
mr_m = mr_m.rename(columns={mr_m.columns[0]: 'Period'})  #name first column as period
mr_q = mr_q.rename(columns={mr_q.columns[0]: 'Period'})   #name first column as period
```

```python
mr_m.iloc[:, 1:] = mr_m.iloc[:, 1:].fillna(method='bfill')  #fill null value to  backward
mr_y.iloc[:, 1:] = mr_y.iloc[:, 1:].fillna(method='bfill')  #fill null value to  backward
mr_q.iloc[:, 1:] = mr_q.iloc[:, 1:].fillna(method='bfill')  #fill null value to  backward
mr_m.iloc[:, 1:] = mr_m.iloc[:, 1:].fillna(method='ffill')  #fill null value to  forward value
mr_y.iloc[:, 1:] = mr_y.iloc[:, 1:].fillna(method='ffill')  #fill null value to  forward value
mr_q.iloc[:, 1:] = mr_q.iloc[:, 1:].fillna(method='ffill')  #fill null value to  forward value
```

#Save all file

```python
#Save cpi file
file_path = 'CPI_DATA.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    # Write each DataFrame to a different worksheet
    cpi_y.to_excel(writer, sheet_name='annual_CPI', index=False)
    cpi_m.to_excel(writer, sheet_name='monthly_CPI', index=False)
```

```python
#save exchange rate file

file_path = 'Exchange_rate.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    # Write each DataFrame to a different worksheet
    ex_y.to_excel(writer, sheet_name='annual_EX', index=False)
    ex_m.to_excel(writer, sheet_name='monthly_EX', index=False)
```

```python
#save export merchandise
file_path = 'Export_Merchandise.xlsx'
with pd.ExcelWriter(file_path, engine='xlsxwriter') as writer:
    # Write each DataFrame to a different worksheet
    mr_y.iloc[1:,:].to_excel(writer, sheet_name='annual_MR', index=False)  #remove  first extra row
    mr_m.to_excel(writer, sheet_name='monthly_MR', index=False)
    mr_q.iloc[1:,:].to_excel(writer, sheet_name='quarterly_MR', index=False) #remove first extra row
```
